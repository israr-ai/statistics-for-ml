```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

# Sample data
data = [2, 3, 3, 4, 4, 4, 5, 6, 6, 7]

# Frequency distribution using Counter
freq = Counter(data)

# Convert to DataFrame for easy plotting
df = pd.DataFrame(list(freq.items()), columns=["Value", "Frequency"]).sort_values("Value")

print("Frequency Table:")
print(df)

# ------------------------------
# 1. Bar Chart
# ------------------------------
plt.figure(figsize=(6,4))
plt.bar(df["Value"], df["Frequency"], color="skyblue", edgecolor="black")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Bar Chart - Frequency Distribution")
plt.show()

# ------------------------------
# 2. Pie Chart
# ------------------------------
plt.figure(figsize=(6,6))
plt.pie(df["Frequency"], labels=df["Value"], autopct="%1.1f%%", startangle=90, colors=plt.cm.Pastel1.colors)
plt.title("Pie Chart - Frequency Distribution")
plt.show()

# ------------------------------
# 3. Histogram
# ------------------------------
plt.figure(figsize=(6,4))
plt.hist(data, bins=range(min(data), max(data)+2), color="lightgreen", edgecolor="black", rwidth=0.8)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram - Frequency Distribution")
plt.show()

# ------------------------------
# 4. Frequency Polygon
# ------------------------------
plt.figure(figsize=(6,4))
plt.plot(df["Value"], df["Frequency"], marker="o", color="red", linestyle="-", linewidth=2)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Frequency Polygon")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
